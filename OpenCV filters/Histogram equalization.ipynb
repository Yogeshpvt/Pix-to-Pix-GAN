```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('Original_3.jpg')
```

```python
col = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(col)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="Greys_r")
```

```python
k = np.asarray(gray)
```

```python
k
```

```python
plt.hist(k.ravel(),256,[0,256])
plt.show()
```

```python
dst = cv2.equalizeHist(k)
```

```python
plt.imshow(dst, cmap='Greys_r')
```

```python
plt.hist(dst.ravel(),256,[0,256])
plt.show()
```

```python
col1 = cv2.cvtColor(dst,cv2.COLOR_GRAY2RGB)
plt.imshow(col1)
```
